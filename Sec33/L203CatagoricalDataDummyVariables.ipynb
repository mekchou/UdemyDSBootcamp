```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv(r"C:\Users\MekChou\OneDrive\Code\Udemy\UdemyDSBootcamp\Data\S33_L203\1.03. Dummies.csv")
raw_data.head()
```

```python
data = raw_data.copy()
data
```

```python
data['Attendance'] = data['Attendance'].map({'Yes': 1,'No': 0})
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(data['SAT'],y)
yhat_no = 0.6439 + 0.0014*data['SAT'] 
yhat_yes = 0.6439 + 0.0014*data['SAT'] + 0.2226
fig = plt.plot(data['SAT'], yhat_no, lw = 2, c = "red")
fig = plt.plot(data['SAT'], yhat_yes, lw = 2, c = "orange")
plt.xlabel ('SAT', fontsize = 20)
plt.ylabel ('GPA', fontsize = 20)
plt.show()
```

```python
plt.scatter(data['SAT'],y,c=data['Attendance'],cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014*data['SAT'] 
yhat_yes = 0.6439 + 0.0014*data['SAT'] + 0.2226
fig = plt.plot(data['SAT'], yhat_no, lw = 2, c = "green")
fig = plt.plot(data['SAT'], yhat_yes, lw = 2, c = "red")
plt.xlabel ('SAT', fontsize = 20)
plt.ylabel ('GPA', fontsize = 20)
plt.show()
```
